```python
# diverse_data_generation.py

import os
import json
import re
import random
import google.generativeai as genai


# API Key 등록
os.environ["GOOGLE_API_KEY"] = "AIzaSyCWEr4TuU58-ITjB-e9ftHjAKBOe2ZBarY"
genai.configure(api_key=os.getenv("AIzaSyCWEr4TuU58-ITjB-e9ftHjAKBOe2ZBarY"))

# 안전 JSON 파서
def safe_json_parse(text: str):
    try:
        return json.loads(text)
    except:
        match = re.search(r"\[.*\]", text, re.S)
        if match:
            return json.loads(match.group(0))
        else:
            print("⚠️ JSON 추출 실패. 원문:", text)
            return []



def normalize_record(record: dict):
    """output JSON 안의 값들을 문자열로 강제 변환"""
    if "output" in record and "actions" in record["output"]:
        for action in record["output"]["actions"]:
            event = action.get("event", {})
            for key in ["date", "time", "title", "location"]:
                if key in event and event[key] is not None:
                    event[key] = str(event[key])  # 무조건 문자열화
            if "participants" in event and isinstance(event["participants"], list):
                event["participants"] = [str(p) for p in event["participants"]]
    return record

# 데이터 증강 함수
def augment_data(seed_example, n=10):
    """
    seed_example: {"input": str, "output": dict}
    n: 생성할 개수
    """
    prompt = f"""
    다음은 일정 관리 예시 데이터입니다.
    입력: {seed_example["input"]}
    출력(JSON): {json.dumps(seed_example["output"], ensure_ascii=False)}

    위 데이터를 참고해서 새로운 학습 데이터를 {n}개 만들어주세요.

    조건:
    - 입력(input)은 다양해야 합니다:
      * 참가자 이름 다양화 (민수, 지영, Emma, Tom 등)
      * 장소 다양화 (헬스장, 카페, 회의실, 공원 등)
      * 활동 다양화 (운동, 회의, 저녁식사, 영화보기, 스터디 등)
      * 시간 표현 다양화 ("내일 7시", "2025-10-05 14:00", "다음주 월요일 아침", "모레 오후 3시")
      * 때때로 일부 정보(장소, 시간 등)가 빠질 수도 있음
    - 액션 다양화:
      * 단순 create
      * 단순 read
      * 단순 update
      * 단순 delete
      * create + read 같이 한 문장에 여러 액션 포함
      * read + update 같이 한 문장에 여러 액션 포함
      * create + update 같이 한 문장에 여러 액션 포함
      * create + delete 같이 한 문장에 여러 액션 포함
    - 출력(JSON)은 반드시 input에 맞게 조정된 올바른 구조여야 합니다.
    - JSONL 학습데이터 형식으로 출력하세요.
      즉, 각 줄마다 {{"input": "...", "output": {{...}}}} 형태로 n개 출력.
    - 불필요한 설명 없이 JSONL만 출력하세요.
    """

    model = genai.GenerativeModel("gemini-2.5-flash")
    response = model.generate_content(prompt)
    text = response.text.strip()

    # 줄 단위 JSONL 파싱
    dataset = []
    for line in text.splitlines():
        parsed = safe_json_parse(line)
        if parsed:
            dataset.append(normalize_record(parsed))  # 문자열 변환 적용
    return dataset

# JSONL 저장
def save_dataset(dataset, filepath):
    with open(filepath, "a", encoding="utf-8") as f:
        for record in dataset:
            f.write(json.dumps(record, ensure_ascii=False, default=str) + "\n")

# 실행 예시
if __name__ == "__main__":
    seed = {
        "input": "내일 저녁 7시에 민수랑 운동 일정 추가해줘",
        "output": {
            "actions": [
                {
                    "action": "create",
                    "event": {
                        "title": "운동",
                        "date": "2025-10-03",
                        "time": "19:00",
                        "participants": ["민수"],
                        "location": "헬스장"
                    }
                }
            ]
        }
    }

    dataset = augment_data(seed, n=10)
    print("생성된 데이터 예시:")
    for d in dataset[:3]:
        print(d)

    SAVE_PATH = "/content/drive/MyDrive/LLM_Schedule/train.jsonl"
    save_dataset(dataset, SAVE_PATH)
    print(f"✅ {SAVE_PATH} 저장 완료!")
```
