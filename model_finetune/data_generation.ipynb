```python
# diverse_data_generation.py

import os
import json
import re
import random
import google.generativeai as genai


# API Key ë“±ë¡
os.environ["GOOGLE_API_KEY"] = "AIzaSyCWEr4TuU58-ITjB-e9ftHjAKBOe2ZBarY"
genai.configure(api_key=os.getenv("AIzaSyCWEr4TuU58-ITjB-e9ftHjAKBOe2ZBarY"))

# ì•ˆì „ JSON íŒŒì„œ
def safe_json_parse(text: str):
    try:
        return json.loads(text)
    except:
        match = re.search(r"\[.*\]", text, re.S)
        if match:
            return json.loads(match.group(0))
        else:
            print("âš ï¸ JSON ì¶”ì¶œ ì‹¤íŒ¨. ì›ë¬¸:", text)
            return []



def normalize_record(record: dict):
    """output JSON ì•ˆì˜ ê°’ë“¤ì„ ë¬¸ìì—´ë¡œ ê°•ì œ ë³€í™˜"""
    if "output" in record and "actions" in record["output"]:
        for action in record["output"]["actions"]:
            event = action.get("event", {})
            for key in ["date", "time", "title", "location"]:
                if key in event and event[key] is not None:
                    event[key] = str(event[key])  # ë¬´ì¡°ê±´ ë¬¸ìì—´í™”
            if "participants" in event and isinstance(event["participants"], list):
                event["participants"] = [str(p) for p in event["participants"]]
    return record

# ë°ì´í„° ì¦ê°• í•¨ìˆ˜
def augment_data(seed_example, n=10):
    """
    seed_example: {"input": str, "output": dict}
    n: ìƒì„±í•  ê°œìˆ˜
    """
    prompt = f"""
    ë‹¤ìŒì€ ì¼ì • ê´€ë¦¬ ì˜ˆì‹œ ë°ì´í„°ì…ë‹ˆë‹¤.
    ì…ë ¥: {seed_example["input"]}
    ì¶œë ¥(JSON): {json.dumps(seed_example["output"], ensure_ascii=False)}

    ìœ„ ë°ì´í„°ë¥¼ ì°¸ê³ í•´ì„œ ìƒˆë¡œìš´ í•™ìŠµ ë°ì´í„°ë¥¼ {n}ê°œ ë§Œë“¤ì–´ì£¼ì„¸ìš”.

    ì¡°ê±´:
    - ì…ë ¥(input)ì€ ë‹¤ì–‘í•´ì•¼ í•©ë‹ˆë‹¤:
      * ì°¸ê°€ì ì´ë¦„ ë‹¤ì–‘í™” (ë¯¼ìˆ˜, ì§€ì˜, Emma, Tom ë“±)
      * ì¥ì†Œ ë‹¤ì–‘í™” (í—¬ìŠ¤ì¥, ì¹´í˜, íšŒì˜ì‹¤, ê³µì› ë“±)
      * í™œë™ ë‹¤ì–‘í™” (ìš´ë™, íšŒì˜, ì €ë…ì‹ì‚¬, ì˜í™”ë³´ê¸°, ìŠ¤í„°ë”” ë“±)
      * ì‹œê°„ í‘œí˜„ ë‹¤ì–‘í™” ("ë‚´ì¼ 7ì‹œ", "2025-10-05 14:00", "ë‹¤ìŒì£¼ ì›”ìš”ì¼ ì•„ì¹¨", "ëª¨ë ˆ ì˜¤í›„ 3ì‹œ")
      * ë•Œë•Œë¡œ ì¼ë¶€ ì •ë³´(ì¥ì†Œ, ì‹œê°„ ë“±)ê°€ ë¹ ì§ˆ ìˆ˜ë„ ìˆìŒ
    - ì•¡ì…˜ ë‹¤ì–‘í™”:
      * ë‹¨ìˆœ create
      * ë‹¨ìˆœ read
      * ë‹¨ìˆœ update
      * ë‹¨ìˆœ delete
      * create + read ê°™ì´ í•œ ë¬¸ì¥ì— ì—¬ëŸ¬ ì•¡ì…˜ í¬í•¨
      * read + update ê°™ì´ í•œ ë¬¸ì¥ì— ì—¬ëŸ¬ ì•¡ì…˜ í¬í•¨
      * create + update ê°™ì´ í•œ ë¬¸ì¥ì— ì—¬ëŸ¬ ì•¡ì…˜ í¬í•¨
      * create + delete ê°™ì´ í•œ ë¬¸ì¥ì— ì—¬ëŸ¬ ì•¡ì…˜ í¬í•¨
    - ì¶œë ¥(JSON)ì€ ë°˜ë“œì‹œ inputì— ë§ê²Œ ì¡°ì •ëœ ì˜¬ë°”ë¥¸ êµ¬ì¡°ì—¬ì•¼ í•©ë‹ˆë‹¤.
    - JSONL í•™ìŠµë°ì´í„° í˜•ì‹ìœ¼ë¡œ ì¶œë ¥í•˜ì„¸ìš”.
      ì¦‰, ê° ì¤„ë§ˆë‹¤ {{"input": "...", "output": {{...}}}} í˜•íƒœë¡œ nê°œ ì¶œë ¥.
    - ë¶ˆí•„ìš”í•œ ì„¤ëª… ì—†ì´ JSONLë§Œ ì¶œë ¥í•˜ì„¸ìš”.
    """

    model = genai.GenerativeModel("gemini-2.5-flash")
    response = model.generate_content(prompt)
    text = response.text.strip()

    # ì¤„ ë‹¨ìœ„ JSONL íŒŒì‹±
    dataset = []
    for line in text.splitlines():
        parsed = safe_json_parse(line)
        if parsed:
            dataset.append(normalize_record(parsed))  # ë¬¸ìì—´ ë³€í™˜ ì ìš©
    return dataset

# JSONL ì €ì¥
def save_dataset(dataset, filepath):
    with open(filepath, "a", encoding="utf-8") as f:
        for record in dataset:
            f.write(json.dumps(record, ensure_ascii=False, default=str) + "\n")

# ì‹¤í–‰ ì˜ˆì‹œ
if __name__ == "__main__":
    seed = {
        "input": "ë‚´ì¼ ì €ë… 7ì‹œì— ë¯¼ìˆ˜ë‘ ìš´ë™ ì¼ì • ì¶”ê°€í•´ì¤˜",
        "output": {
            "actions": [
                {
                    "action": "create",
                    "event": {
                        "title": "ìš´ë™",
                        "date": "2025-10-03",
                        "time": "19:00",
                        "participants": ["ë¯¼ìˆ˜"],
                        "location": "í—¬ìŠ¤ì¥"
                    }
                }
            ]
        }
    }

    dataset = augment_data(seed, n=10)
    print("ìƒì„±ëœ ë°ì´í„° ì˜ˆì‹œ:")
    for d in dataset[:3]:
        print(d)

    SAVE_PATH = "/content/drive/MyDrive/LLM_Schedule/train.jsonl"
    save_dataset(dataset, SAVE_PATH)
    print(f"âœ… {SAVE_PATH} ì €ì¥ ì™„ë£Œ!")
```
